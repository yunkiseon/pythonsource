#### 정규표현식
- 정규식 패턴은 동일

```python
data = """ \
park 800905-1049118
kim 700905-1059118
"""

# 뒷자리를 ****** 출력

# 엔터를 기준으로 문자열 분리
words = data.split("\n")
print(words)

# 결과를 담을 리스트 생성
result = []

# 공백을 기준으로 문자열 분리
for line in words:
    word_result = []
    # person = line.strip().split(" ")
    for word in line.strip().split(" "):
        if len(word) == 14 and word[:6].isdigit() and word[7:].isdigit():
            word = word[:6] + "-" + "*******"
        word_result.append(word)

    result.append(" ".join(word_result))

# print(result)

# 리스트를 문자열로 
print("\n".join(result))
```

#### 정규식
- 1. 문자열이 가지고 있는 메소드 사용
- 2. re 모듈
    - compile() : 패턴 생성
    - search() : 문자열 전체를 검색해서 정규식과 매칭되는 패턴을 찾아서 리턴(Match 객체)
    - sub() : 찾아서 변경
    - findall() : 패턴과 일치하는 모든 문자열을 찾아서 리스트로 반환

```python
import re
from openpyxl import load_workbook, Workbook
import requests
```

```python
# 패턴
# . , *, +, ?,...

# 패턴 지정 : . (모든 것 의미)
pattern = re.compile("D.A")
result = pattern.search("DAA")
print(result)

# Match object 접근
print("패턴 시작 위치",result.start())
print("패턴 종료 위치",result.end())
print("패턴과 일치하는 문자열 반환",result.group())
print("패턴 시작, 종료 위치",result.span())
```

```python
result = pattern.search("d0A D1A 0111")
print("패턴 시작 위치",result.start())
print("패턴 종료 위치",result.end())
print("패턴과 일치하는 문자열 반환",result.group())
print("패턴 시작, 종료 위치",result.span())
```

```python
# *, +, ?, {3}, {3,5} : 숫자 지정

# ? : 0 ~ 1
pattern = re.compile("D?A")  # DA
print(pattern.search("A"))
print(pattern.search("DA"))
print(pattern.search("AA"))

# + : 1 ~ 무한대
pattern = re.compile("D+A")  # DA
print(pattern.search("A"))
print(pattern.search("DA"))
print(pattern.search("AA"))


# * : 0 ~ 무한대
pattern = re.compile("D*A")  # DA
print(pattern.search("A"))
print(pattern.search("DA"))
print(pattern.search("DDDDDDDDDDDDDDDDDDDDDDDDDDDDDDDDDDDDDAA"))

# {최소,최대}
# {3}
pattern = re.compile("D{3}A")  # DA
print(pattern.search("A"))
print(pattern.search("DA"))
print(pattern.search("DDDDDDDDDDDDDDDDDDDDDDDDDDDDDDDDDDDDDAA"))

pattern = re.compile("D{3,10}A")  # DA
print(pattern.search("A"))
print(pattern.search("DA"))
print(pattern.search("DDDDDDDDDDDDDDDDDDDDDDDDDDDDDDDDDDDDDAA"))
```

```python
# train 읽어서 남성,미혼여성,기혼여성,기타 각각의 시트로 저장
# 보고서 시트 생성
# 분류  생존자수 사망자수 생존률
# 남성    150    80     60

wb = Workbook()
ws_man = wb.active
# 시트명 변경
ws_man.title = "남성"

ws_solo_women = wb.create_sheet()
ws_solo_women.title = "미혼여성"

ws_married_women = wb.create_sheet()
ws_married_women.title = "기혼여성"

ws_others_women = wb.create_sheet()
ws_others_women.title = "기타"

ws_report = wb.create_sheet(title="보고서")

# 생존자수, 사망자수 변수 선언
man_survived, man_unsurvived = 0, 0
women_survived, women_unsurvived = 0, 0
solo_women_survived, solo_women_unsurvived = 0, 0
others_survived, others_unsurvived = 0, 0


# 남자( Mr\.),미혼여성( Miss\.), 기혼여성( Mrs\.), 기타
pattern = re.compile(r" [A-Za-z]+\.")

# 엑셀 데이터 가져오기
train_wb = load_workbook("./data/train.xlsx")
train_ws = train_wb.active

for row in train_ws.iter_rows():
    # 이름 컬럼
    result = pattern.search(row[3].value)
    # 생존 여부 컬럼
    survived = row[1].value

    # 제목 행 붙여넣기
    if row[0].row == 1:
        ws_man.append([item.value for item in row])
        ws_married_women.append([item.value for item in row])
        ws_solo_women.append([item.value for item in row])
        ws_others_women.append([item.value for item in row])
    else:
        if result:
            name = result.group()
            if name.strip() == "Mr.":
                ws_man.append([item.value for item in row])
                # servived 0 인 경우 남자사망자수 += 1
                if survived == 0:
                    man_unsurvived += 1
                else:
                # servived 1 인 경우 남자생존자수 += 1
                    man_survived += 1
            elif name.strip() == "Miss.":
                ws_solo_women.append([item.value for item in row])
                if survived == 0:
                    solo_women_unsurvived += 1
                else:               
                    solo_women_survived += 1
            elif name.strip() == "Mrs.":
                ws_married_women.append([item.value for item in row])
                if survived == 0:
                    women_unsurvived += 1
                else:               
                    women_survived += 1
            else:
                ws_others_women.append([item.value for item in row])            
                if survived == 0:
                    others_unsurvived += 1
                else:               
                    others_survived += 1

# 보고서 작성
ws_report.append(["분류","생존자수", "사망자수", "생존률"])

rate = "%.2f%%" % (man_survived / (man_survived + man_unsurvived) * 100)
ws_report.append(["남성",man_survived, man_unsurvived, rate])

rate = women_survived / (women_survived + women_unsurvived) * 100
ws_report.append(["기혼여성",women_survived, women_unsurvived, rate])

rate = solo_women_survived / (solo_women_survived + solo_women_unsurvived) * 100
ws_report.append(["미혼여성",solo_women_survived, solo_women_unsurvived, rate])

rate = others_survived / (others_survived + others_unsurvived) * 100
ws_report.append(["기타",others_survived, others_unsurvived, rate])

wb.save("./data/train_gender.xlsx")
wb.close()
```
