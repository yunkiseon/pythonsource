```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import urllib.request as req
from selenium.common.exceptions import NoSuchElementException
```

```python
# 브라우저 창 안보여주는 옵션
options = webdriver.ChromeOptions()
options.add_argument("--headless")
driver = webdriver.Chrome(options=options)
driver.maximize_window()

driver.get("http://www.google.com")

print(f"title {driver.title}")
print(f"url {driver.current_url}")
print(f"source {driver.page_source}")

time.sleep(3)
driver.close()
```

```python
# iframe 으로 구성된 페이지 접근


driver = webdriver.Chrome()
driver.maximize_window()

driver.get("https://www.w3schools.com/html/tryit.asp?filename=tryhtml_basic_document")

time.sleep(3)

# iframe 안으로 이동
driver.switch_to.frame("iframeResult")

# iframe 내부 요소 찾기
element = driver.find_element(By.TAG_NAME, "h1")
print(element.text)

# iframe 밖으로 나오기
driver.switch_to.default_content()

# 왼쪽 요소 찾기
# #textareawrapper > div > div.CodeMirror-scroll > div.CodeMirror-sizer > div > div > div > div.CodeMirror-code > pre:nth-child(7)
pre = driver.find_element(By.CSS_SELECTOR, "#textareawrapper > div > div.CodeMirror-scroll > div.CodeMirror-sizer > div > div > div > div.CodeMirror-code > pre:nth-child(7)")
print(pre.text)

time.sleep(2)

# 4. 파싱 : selenium or BeautifulSoup
titles = driver.find_elements(By.TAG_NAME, "h3")
for title in titles:
    print(title.text)

driver.close()
```

```python
driver = webdriver.Chrome()
driver.maximize_window()

driver.get("https://www.naver.com/")

time.sleep(3)

# 검색창에 검색어 입력
# 1. 검색창 찾기
element = driver.find_element(By.ID, "query")
# 2. 검색어 입력
element.send_keys("바이올린")
element.send_keys(Keys.ENTER)

# 3. 결과페이지로 이동할 때까지 시간 부여
time.sleep(2)

# 4. 파싱 : selenium or BeautifulSoup

## 엑스패스 : //*[@id="lnb"]/div[1]/div/div[1]/div/div[1]/div[2]/a
# 이미지 클릭
driver.find_element(By.XPATH,'//*[@id="lnb"]/div[1]/div/div[1]/div/div[1]/div[1]/a').click()
time.sleep(2)

# 현재 문서 높이
prev_height = driver.execute_script("return document.body.scrollHeight")

while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")

    time.sleep(2)

    # 스크롤이 진행된 후 문서 높이
    curr_height = driver.execute_script("return document.body.scrollHeight")

    if curr_height == prev_height:
        break

    prev_height = curr_height

# 스크롤 이동 완료 후 맨 처음으로 스크롤 이동
driver.execute_script('window.scrollTo(0,0)')

time.sleep(3)

# 이미지 저장
# 작은 썸네일 이미지 전체 가져오기
images = driver.find_elements(By.CSS_SELECTOR,".thumb .link_thumb")

count =1 
for img in images:
    img.click()
    time.sleep(2)
    # 오른쪽에 나오는 큰 이미지 저장
    # div.viewer_image img > img
    large_image = driver.find_element(By.CSS_SELECTOR,"div.viewer_image img").get_attribute("src")
    req.urlretrieve(large_image,f"./download/{count}.png")
    count +=1

driver.close()
```

```python
# 페이징된  사이트 크롤링
url = "https://prod.danawa.com/list/?cate=112758&15main_11_02="

driver = webdriver.Chrome()
driver.maximize_window()

driver.get(url)

time.sleep(3)

# 광고창 클릭
# //*[@id="newProductBanner"]/div/div/div[3]/button[2]/span
driver.find_element(By.XPATH,'//*[@id="newProductBanner"]/div/div/div[3]/button[2]/span').click()

time.sleep(3)
# M4 맥북 클릭
# //*[@id="viewMore_43778"]
driver.find_element(By.XPATH,'//*[@id="viewMore_43778"]').click()
time.sleep(3)


cur_page, target_crawl_num = 1,4
idx = 1
while cur_page <= target_crawl_num:
    print(f"현재 페이지 : {cur_page}")
    # 제품명, 이미지, 가격(최저가)
    products = driver.find_elements(By.CSS_SELECTOR,".product_list .prod_item .prod_main_info")
    
    for product in products:
        product_name = product.find_element(By.CSS_SELECTOR,".prod_name a").text
        product_image = product.find_element(By.CSS_SELECTOR,".thumb_image img")
        if product_image.get_attribute("src"):
            src = product_image.get_attribute('src')
        else:
            src = product_image.get_attribute('data-original')
        
        if "https:" not in src:
            src = "https:"+src
        
        try:
            price = product.find_element(By.CSS_SELECTOR,".price_sect strong").text
        except NoSuchElementException:
            price = 0
    


        print(product_name,src, price)
        idx += 1

    cur_page += 1
    if cur_page > target_crawl_num:
        break
    # 페이지 번호 클릭
    driver.find_element(By.CSS_SELECTOR,f"div.number_wrap a:nth-child({cur_page})").click()
    time.sleep(2)




driver.close()
```
