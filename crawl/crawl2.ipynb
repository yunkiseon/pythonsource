```python
from fake_useragent import UserAgent
import requests
from bs4 import BeautifulSoup
import csv
import pymysql
from pymongo import MongoClient
```

```python
url = "https://openapi.naver.com/v1/search/book.json"

headers = {
    "X-Naver-Client-Id" : "WsDSDU4TWHkbKVF6MHSP",
    "X-Naver-Client-Secret" : "Ye1bBpI6Ze"
}


start = 1

# csv 저장
books = []
no =1

# sql 저장
books_sql = []

# mongo 저장
books_mongo = []

# book -> title, author, discount, publisher

with requests.Session() as s:
    # range -> 0~9    
    for idx in range(3):
        start_num = start + (idx * 100)
        param = {
        "query" : "고양이",
        "display" : "10",
        "start" : start_num
        }
        res = s.get(url,headers=headers, params=param)
        contents = res.json()
        for book in contents['items']:
            books.append((no,book['title'],book['author'],book['discount'],book['publisher']))
            books_sql.append((book['title'],book['author'],book['discount'],book['publisher']))
            book_m = {
                "title":book['title'],
                "author":book['author'],
                "discount":book['discount'],
                "publisher":book['publisher']
            }
            books_mongo.append(book_m)
            no +=1
```

```python
# 로컬 저장
with open("./data/books.csv","w",newline="",encoding="utf-8")as fw:
    wt = csv.writer(fw)
    for v in books:
        wt.writerow(v)
```

```python
# db 입력
with conn.cursor()as cursor:
    sql = "insert into books(title, author, discount, publisher) values(%s,%s,%s,%s)"
    cursor.executemany(sql,books_sql)
conn.commit()
```
