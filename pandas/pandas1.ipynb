```python
import pandas as pd
```

```python
series = pd.Series([70,60,90])
series
```

```python
series = pd.Series({"a":1,"b":2,"c":3})
series
```

```python
series = pd.Series([70,60,90],index=["국어","영어","수학"])
series
```

```python
# 조회
print(series.국어)
print(series.index)
print(series.values)
print(series.shape) # 튜플로 몇 행인지
print(series.head(1)) # 앞에서부터  하나
print(series.tail(1)) # 뒤에서부터  하나
```

```python
# 수정
series.index = ["미술","음악","체육"]
series
# dtype: int64 -> int 64비트로 표현하는 중
# java : int 4 byte, long : 8 byte

# 타입변경
series = series.astype("int32")
series
```

```python
# 제거
del series['음악']
series
```
