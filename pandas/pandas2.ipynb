```python
# null? not null?
df = pd.DataFrame({"a":[4,5,6,6,],"b":[7,8,9,9],"c":[10,11,12,12]},index=['a','b','c','d'])
df
```

```python
# b 값이 7이 아닌
df[df.b != 7]
```

```python
# b 값이 7이 아닌 / a,b 칼럼만 조회
df.loc[df.b != 7, "a":"b"]
```

```python
# a 값에 5가 들어있는지 확인 -> df.a.isin([5]) 는 true, false 반환
df[df.a.isin([5])]
```

```python
# #isna(), isnull() : true, false 반환
df.isna()
df.a.isnull()
```

```python
df = sns.load_dataset("mpg")
df
```

```python
# group by
df.groupby("origin").count()
df.groupby("origin").size()
df.groupby("origin")['weight'].mean()
df.groupby(["origin","model_year"])['cylinders'].mean()
df.groupby("origin")['horsepower'].agg(['sum','mean','std'])
```
