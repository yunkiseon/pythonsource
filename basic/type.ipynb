```python
# 입력 -> 년/월/일 (2026/02/26)
# 10년 후의 날짜를 계산 후 보여주기 => 2036년 02월 26일
origin = input('년/월/일')
orisp = origin.split('/')
year = int(orisp[0]) + 10
month = orisp[1]
day = orisp[2]
print(f'{year}년 {month}월 {day}일')
```

```python
# 비밀번호 생성
# http://naver.com이라고 할 대
# 규칙 1. http:// 제외-> naver.com
# 규칙 2. 첫번째로 나오는 마침표 이후 부분 제외 -> naver
# 규칙 3. 남은 글자 중 첫 세자리 + 글자 개수 + 글자 내 'e' 문자 개수 + !

password = "http://naver.com" 
password = password[7:]
# origin.replace("http://","") 해서 잘라내기
password = password[0:5]
# ori1[:ori1.find(".")]로 대체 가능
password = password[0:3] + str(len(password)) + str(password.count('e')) + "!"
print(password)
```

```python
# 100 이상의 숫자 개수
numbers = [273,103,5,32,65,9,72,800,99]
count = 0
for i in range(len(numbers)):
    A = numbers[i]
    if A >= 100:
        print(A)
        count += 1
print(f"100이상 수 {count} 개")
```

```python
# numbers의 숫자와 자릿수 출력
# 출력예시 : 273 -> 3자릿수
for i in range(len(numbers)):
    A = str(numbers[i])
    print(f'숫자 : {A}, 자릿수 :{len(A)}')
```
