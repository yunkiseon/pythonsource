````python
# 주차장 프로그램
# 5대만 주차 가능 (차례대로 입차/출차는 반대 순서)
parking_lot = []
top, car_name = 0, "A"
while True:
    no = int(input('[1] 자동차 넣기 | [2] 자동차 빼기 | [3] 종료 '))
    if no <= 3:
        if no == 1:
            if top >= 5:
                print("만차입니다.")
            else:
                parking_lot.append(car_name)
                print(f"{car_name} 자동차 들어감. 주차장 상태 ==> {parking_lot}")
                top += 1
                # car_name 새로 구하기 
                car_name = chr(ord(car_name) + 1)
        if no == 2:
            if top > 0:
                # 마지막 꺼내기: pop()
                parking_lot.pop()
                print(f'{car_name} 자동차 나감. 주차장 상태 ==> {parking_lot}')
                top -= 1
                car_name = chr(ord(car_name)-1)
            else : 
                print("빠져나갈 자동차가 없음")
        if no == 3:
            print("프로그램 종료")
            break
    else:
        print("메뉴 번호를 확인해 주세요.")
````

````python
def sum_mul(choice, *args):
    if choice == "+":
        result = 0
        for i in args:
            result += i
    elif choice == "*":
        result = 1
        for i in args:
            result *= i
    return result
print(sum_mul("+",1,2,3,4,5))
print(sum_mul("*",1,2,3,4,5))
````

````python
# 함수 정의 : 사칙연산()+-*/
def fourRules(num1,num2,op):
    if op =="+":
        print(num1 + num2)
    elif op =="-":
        print(num1 - num2)
    elif op =="*":
        print(num1 * num2)
    elif op =="//":
        print(num1 // num2)

fourRules(10,30,"//")

## 사용자한테 두 수와 연산자 입력받기
num1 = int(input("num1 입력"))
num2 = int(input("num2 입력"))
op = input("op 입력")
fourRules(num1,num2,op)
````

````python
# 숫자를 받아서 **2 리턴
def square(num):
    return num ** 2
print(square(5))
````

````python
square = lambda num:num **2
square(6)
````

````python
add = lambda x,y:x+y
add(6,9)
````

````python
# def strlen(s):
#     return len(s)

content = ['bob','charles','alexander3','teddy']

# 오름, 내림의 sort가 아닌 글자 길이에 따른 sort
# content.sort()
# content.sort(reverse=True)

content.sort(key=lambda s:len(s))
print(content)
````

````python
content = ['bob','charles','alexander3','teddy']
content.sort(key=len)
print(content)
````

````python
# filter, map, reduce
# filter -> 조건 만족하는 요소들만 새로운 list로
list1 = [1,2,3,4,5,8,9]
# 짝수요소만 추출후 새로운 리스트 생성
list2 = [i for i in list1 if i % 2 == 0]
list2
````

````python
evens = []
def even(list1):
    for i in list1:
        if i % 2 == 0:
            evens.append(i)
even([1,5,8,96,3,26,7,4,6,8])
print(evens)
````

````python
# filter 사용
list1 = [1,2,3,4,6,8,9]
list(filter(lambda x:x%2==0, list1))
````

````python
# map
nums = [1,2,3,5,6,10,11,13,15,16,19]
list(map(lambda x:x**2,nums))
````
