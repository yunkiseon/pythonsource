```python
# 1000 명의 키, 몸무게 임의 생성 후 파일 저장
import random

# list(range(1,101,2))
# list("abcedtfbg")
hanguls = list("가나다라마바사아자차카타파하")

fw = open("info.txt", "w", encoding="utf-8")
for i in range(1000):
    name=random.choice(hanguls) + random.choice(hanguls)
    weight = random.randrange(40,100)
    height = random.randrange(140,200)
    fw.writelines(f"{name}, {weight},{height}\n")
fw.close()
```

```python
# score.txt 읽기
# 총점, 평균 구하기
fr = open("score.txt", "r", encoding="utf-8")
scores = fr.readlines()
scores = list(map(int,scores))
print(f"총점 : {sum(scores)} 평균 : {sum(scores)/len(scores):.2f}")
fr.close()
```

```python
# origin 읽기 -> 암호화시킨 후 저장 
while True:
    no = int(input("1.암호화 | 2.복호화 | 3.종료 중 원하는 메뉴를 입력해주세요."))
    if no == 1:
        # origin.txt 읽기 -> 암호화 후 저장(encryption.txt) 암호화조건: ord +100
        fr = open("origin.txt", "r", encoding="utf-8")
        content = fr.read()
        fw = open("encryption.txt","w",encoding="utf-8")
        for c in content:
            fw.write(chr(ord(c)+100))
        fr.close
        fw.close()
# 암호화파일을 읽고 그 이후 내용복원시키고 출력
    elif no ==2:
        fr = open("encryption.txt", "r", encoding="utf-8")
        content = fr.read()
        for i in range(len(content)):
            print(chr(ord(content[i])-100),end="")
        fr.close
    else:
        break
```
