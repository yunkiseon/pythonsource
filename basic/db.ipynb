#### 연결
- connect() : db 서버 연동
- cursor : DB 와 연결 후 접근할 수 있는 객체

```python
import pymysql
```

```python
conn = pymysql.connect(host="localhost",user="javadb",password='12345',database='springdb',charset="utf8mb4")
```

#### 조회
- fetchall()
- fetchone()
- fetchmany()

```python
with conn.cursor() as cursor:
    sql = "select * from booktbl"
    cursor.execute(sql) # sql 구문 실행하라는 의미
    rows = cursor.fetchall()
    for row in rows:
        print(row)
```

```python
with conn.cursor() as cursor:
    sql = "select * from booktbl"
    cursor.execute(sql) # sql 구문 실행하라는 의미
    rows = cursor.fetchmany(10)
    for row in rows:
        print(row)
```

```python
with conn.cursor() as cursor:
    sql = "select * from booktbl"
    cursor.execute(sql) # sql 구문 실행하라는 의미
    rows = cursor.fetchone()
    print(rows)
```

```python
with conn.cursor() as  cursor:
    sql = "create table if not exists users(id int primary key auto_increment,name varchar(30),"\
    "email varchar(100), phone varchar(20), website varchar(50), regdate date)"
    cursor.execute(sql)
```

```python
# 날짜 모듈
# import datetime
# datetime.datetime.now()
from datetime import datetime
datetime.now()#datetime.datetime(2026, 2, 9, 16, 46, 51, 488280)
# format 지정
now = datetime.now()
today = now.strftime("%Y-%m-%d")
```

```python
# 삽입-튜플 구조로 밀어넣기
with conn.cursor()as cursor:
    sql = "insert into users(name, email, phone, website, regdate) values(%s,%s,%s,%s,%s)"
    cursor.execute(sql,('user1','user1@gmail.com','010-1234-5678','http://user1.com',today))
conn.commit()
```

```python
with conn.cursor() as cursor:
    sql = "select * from users"
    cursor.execute(sql) 
    rows = cursor.fetchmany()
    for row in rows:
        print(row)
```

```python
with conn.cursor()as cursor:
    users = (
        ('user2','user2@gmail.com','010-1234-5678','http://user2.com',today),
        ('user3','user3@gmail.com','010-1234-5678','http://user3.com',today),
        ('user4','user4@gmail.com','010-1234-5678','http://user4.com',today),
        ('user5','user5@gmail.com','010-1234-5678','http://user5.com',today)
    )
    sql = "insert into users(name, email, phone, website, regdate) values(%s,%s,%s,%s,%s)"
    cursor.executemany(sql,users)
conn.commit()
```

```python
with conn.cursor() as cursor:
    sql = "select * from users where id=3"
    cursor.execute(sql) 
    rows = cursor.fetchmany()
    for row in rows:
        print(row)
```

```python
with conn.cursor() as cursor:
    sql = "select * from users where id=%s"
    cursor.execute(sql,5) 
    rows = cursor.fetchmany()
    for row in rows:
        print(row)
```

```python
with conn.cursor() as cursor:
    sql = "select * from users where id in (%s,%s)"
    cursor.execute(sql,(3,5)) 
    rows = cursor.fetchall()
    for row in rows:
        print(row)
```

```python
# 수정
with conn.cursor() as cursor:
    sql = "update users set name=%s where id=%s"
    cursor.execute(sql,('change2',2))
conn.commit()
```

```python
# 삭제
with conn.cursor() as cursor:
    sql = "delete from users where id =%s"
    cursor.execute(sql,5)
conn.commit()
```
