- word.txt 읽은 후 리스트 생성
- 리스트 섞은 후 단어를 하나 추출 -> 추출된 단어 보여주기
- 사용자는 보여진 단어를 똑같이 타자입력하기
- 5회 반복하여 제출하기
    - bugs 에 오타를 모으고 정확히 입력한 단어의 개수가 3개 초과라면 합격/불합격
- 게임시간 재기

- 출력문 예제
- 게임시간 : 30.05초, 정답개수 : 4개 
- 합격

```python
import time
import random
import pymysql
```

```python
conn = pymysql.connect(host="localhost",user="javadb",password='12345',database='springdb',charset="utf8mb4")

with conn.cursor() as  cursor:
    sql = "create table if not exists records(id int primary key auto_increment,"\
    "correct_answer int, recore decimal(10,2),regdate date default (curdate()))"
    cursor.execute(sql)
```

```python
with open("./data/word.txt","r",encoding="utf-8") as fr:
    start = time.time()
    content = list(fr.readlines())
    count = 0
    for i in range(1,6):
        random.shuffle(content)
        Q = content[0].strip()
        A = input(f"{Q}")
        if Q == A:
            count += 1
        else:
            count += 0
    end = time.time()
    long = end - start
    if count >= 3:
        print(f"게임시간 {long:.2f}, 정답개수:{count}, 합격입니다")
    else:
        print(f"게임시간 {long:.2f}, 정답개수:{count}, 불합격입니다")
    
    with conn.cursor()as cursor:
        sql = "insert into records(correct_answer, recore) values(%s,%s)"
        cursor.execute(sql,(count,long))
conn.commit()

fr.close()
```
