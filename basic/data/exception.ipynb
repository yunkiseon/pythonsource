```python
try : 
    에러가 발생할 수 있는 구문
except:
    에러 발생 시 해야할 일
else:
    에러가 나지 않는 경우 처리 구문
finally:
    에러 발생여부와 상관없이 처리 구문
```

```python
try:
    a = 10
    b =10
    print(c)
except NameError:
    print("변수명 확인")
```

```python
try:
    x = [10,20,30]
    print(x[3])
except IndexError as e:
    print(e)
```

```python
try:
    dic1 = {"name":"hong","age":33}
    print(dic1["hobby"])
except KeyError as e:
    print(e)
except Exception as e:
    print(e)
```

```python
name = ["Kim","park","Lee"]
try:
    z = "Lee"
    x = name.index(z)
except:
    print("이름이 없습니다.")
else:
    print("이름을 찾았습니다 ", x)
```

```python
# 예외 직접 발생 : 자바 -> throw new Exception()
# 파이썬에선 raise 사용
try:
    a='333'
    if a == 'Kim':
        print("Ok")
    else:
        raise ValueError
except ValueError:
    print("에러")
else:
    print(a)
```
