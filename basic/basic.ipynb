```python
# 사칙연산(숫자 2, 연산자)
num1 = int(input("첫번째 숫자"))
num2 = int(input("두번째 숫자"))
f = str(input())

if f == '+':
    print(num1 + num2)
elif f == '-':
    print(num1 - num2)
elif f == '*':
    print(num1 * num2)
elif f == '/':
    print(num1 / num2)
elif f == '//':
    print(num1 // num2)
elif f == '%':
    print(num1 % num2)
```

```python
# 1 ~100 합계
i = 1
total = 0
while i < 101:
    i += 1
    total += i
print(f"1~100까지의 합 {total}")
```

```python
#'q' 문자가 입력되기 전까지 사용자 입력받기
content = ""
while content != 'q':
    content = input("입력을 중지하고 싶다면 q >>")
    print(content)
```
