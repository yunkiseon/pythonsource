```python
from pymongo import MongoClient
from bson import ObjectId
```

```python
# 연결
client = MongoClient("mongodb+srv://yunkiseon:qweiopasd123@cluster0.koeeh90.mongodb.net/?appName=Cluster0")
db = client["sample_mflix"]
```

```python
# 전체 collection 가져오기
collections = db.list_collection_names()
for collection in collections:
    print(collection)
```

```python
# 요소가져오기(제한걸기)
docs = db.movies.find({"rated":"TV-G"}).limit(10)
for doc in docs:
    print(doc)
```

```python
# 요소가져오기(정렬) sort({필드명} : 1) 오름차순
# find({}),find(): 전체 / find({},{컬럼명}) : 전체조회하되 컬럼은 지정한 것만
docs = db.movies.find({},{"title":1,"imdb.rating":1,"_id":0}).sort({"year":-1}).limit(10) # -1 은 내림차순
for doc in docs:
    print(doc)
```

```python
# 상영시간 100분 이상인 영화 조회(제목, 상영시간, 등급)
docs = db.movies.find({"runtime":{"$gte":100}},{"title":1,"runtime":1,"rated":1,"_id":0}).limit(30)
for doc in docs:
    print(doc)
```

```python
# imdb 평점이 8점 이상 영화제목 (제목, imdb) 10개만
docs = db.movies.find({"imdb.rating":{"$gte":8}},{"title":1,"imdb":1,"_id":0}).limit(10)
for doc in docs:
    print(doc)  
```

```python
# imdb 평점이 8점 이상이고 2000년 이후 영화제목 (제목, imdb, year) 10개만
docs = db.movies.find({"imdb.rating":{"$gte":8},"year":{"$gte":2000}},{"title":1,"imdb":1,"year":1,"_id":0}).limit(10)
for doc in docs:
    print(doc)  
```

```python
# 제목에 love 대소문자 구분없이 들간 영화의 plot, title 10개만
docs = db.movies.find({"title":{"$regex":"love","$options":"i"}},{"title":1,"plot":1,"_id":0}).limit(10)
for doc in docs:
    print(doc)  
```

```python
# 제작연도가 1920년 이전인 영화 개수
# len은 안됨 커서타입이어서
# 그래서 len(docs.to_list()) 필요함 혹은 아래로.
print(db.movies.count_documents({"year":{"lt":1920}}))
print(db.movies.count_documents({"genres":"Comedy"}))
docs = db.movies.find({"year":{"lt":1920}})
count = 0
for doc in docs:
    count += 1
print(count)
```

```python
#  장르가 comedy, romance 둘 다 포함된 영화, title, genres, 10개
# in = or 
# 그래서 $all 사용
docs = db.movies.find({"genres":{"$all":["Comedy","Romance"]}},{"title":1,"genres":1,"_id":0}).limit(10)
for doc in docs:
    print(doc)
```

```python
# 최신영화 10 개 sort-year
docs = db.movies.find().sort({"year":-1}).limit(10)
for doc in docs:
    print(doc)
```

```python
# 영화 제작 년도가 1990~19999 사이의 영화 조회 title, year 10개
docs = db.movies.find({"year":{"$gte":1990},"year":{"$lte":1999}},{"title":1,"year":1,"_id":0}).limit(10)
for doc in docs:
    # print(doc)
    print(doc["title"],doc['year'])
```

```python
# 모든 장르 보기
print(db.movies.distinct("genres"))
```

```python
# 연도별 영화개수 
# 그룹바이-년도, 카운트 해야할 것을 아래 방식으로 실행
rows = db.movies.aggregate([
                    {"$group":{"_id":"$year","count":{"$sum":1}}},
                    {"$sort":{"_id":-1}},
                    {"$limit":10}])
for row in rows:
    print(row)
```

```python
# 장르별 영화개수
rows = db.movies.aggregate([
                    {"$unwind":"$genres"},
                    {"$group":{"_id":"$genres","count":{"$sum":1}}},
                    {"$sort":{"_id":-1}},
                    {"$limit":10}])
for row in rows:
    print(row)
```

```python
# imdb 평점 평균이 높은 감독
rows = db.movies.aggregate([
                    {"$unwind":"$directors"},
                    {"$group":{"_id":"$directors","avgRating":{"$avg":"$imdb.rating"},"count":{"$sum":1}}},
                    {"$match":{"count":{"$gte":5}}},
                    {"$sort":{"avgRating":-1}},
                    {"$limit":10}])
for row in rows:
    print(row)
```
