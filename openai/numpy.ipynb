#### 넘파이(numpy)
- pandas, 머신러닝 라이브러리에서 기본적으로 필요한 라이브러리

```python
import numpy as np
```

```python
# 1차원 백터 생성
array1 = np.array([1,2,3])
array1 #array([1, 2, 3])
```

```python
print(type(array1)) # numpy.ndarray
print(array1.shape) #(3,)
```

```python
# 2 차원 백터 생성
array2 = np.array([[1,2,3],[4,5,6]])
print(array2)
print(array2.shape) # (2,3)
```

```python
# 파이썬 리스트
list1 = [1,2,'a','b',True]
list1
```

```python
array1.dtype # int64
```

```python
array3 = np.array([1,2,3.1])
print(array3.dtype) # float64 -> 큰수에 맞춰줌
```

```python
array4 = np.array(list1)
print(array4.dtype)
```

```python
# 행렬 사칙연산
x = np.array([[4,4,4],[8,8,8]])
y = np.array([[1,1,1],[2,2,2]])
```

```python
print(x+y)
print(x-y)
print(x*y)
print(x//y)
```

```python
x = np.array([[1,2,3],[4,5,6]])
y = np.array([[7,8],[9,10],[11,12]])
```

```python
print(np.dot(x,y))
```

```python
# 전치행렬
np.transpose(x)
```

```python
v = np.array([3,4])

# norm(ord=Norm 의 종류)
# L1(Manhattan distance), L2(Euclidean distance), Max(절대값 중 가장 큰 값)


np.linalg.norm(v)
```
