```python
from dotenv import load_dotenv, find_dotenv
load_dotenv(find_dotenv()) # read local .env file
from openai import OpenAI

# 유튜브 자막 추출
from youtube_transcript_api import YouTubeTranscriptApi
```

##### 유튜브 자막 추출 후 핵심 내용 요약(자막 추출 -> LLM 호출 -> 응답)

```python
client = OpenAI()
model = "gpt-5-nano"
```

```python
# OpenAI CEO 연설(7:16)
# https://www.youtube.com/watch?v=Pn-W41hC764  

# ebs 다큐(38:18)
# https://www.youtube.com/watch?v=e8cpV870Ufs

video_id = "Pn-W41hC764"  
ytt_api = YouTubeTranscriptApi()
# 자막 가져오기
transcript= ytt_api.fetch(video_id,languages=["en"])

# 자막을 텍스트로 정리
text = " ".join([t.text for t in transcript])

#--------------------------------------------------------
# 영어로 추출된 내용을 한국어로 번역 후 내용 요약 시키기
#--------------------------------------------------------

# OpenAI로 요약

response = client.responses.create(
    model=model,
    instructions="""
    넌 영어 자막을 한국어로 번역하고, 핵심 내용을 요약하는 전문가다.
    자막이 영어인 경우 자연스러운 한국어로 변역한 뒤, 주요 내용을 10 줄 이내로 요약해.
    요약은 * 불릭 포인트 형식으로 작성해
    """,    
    input=text
)
print(response.output_text)
```
